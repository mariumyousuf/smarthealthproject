### Pressure sensor plots

This file contains plots for the 11 patients from the ground truth data for which pressure sensor data is available. The other 10 patients were either not found, or were missing the pressure sensor data. 

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os 
import csv
import scipy as sp
import re
from sklearn import preprocessing
from mpl_toolkits.mplot3d import Axes3D
#plt.style.use('seaborn')
```

```python
#pressure_sensors = []
for filename in os.listdir(os.getcwd()):
    if len(filename) == 8:
        print(filename)
        os.chdir(filename)
        for file in os.listdir(os.getcwd()):
            print("file: ", file)
            if file.startswith("9SensChairData"):
                pressure_data = pd.read_csv(file, sep=",", header=None)
                #pressure_data.head()      
                pressure_data.columns = ['time_stamp','sensor_1', 'sensor_2', 'sensor_3', 'sensor_4', 
                                         'sensor_5', 'sensor_6', 'sensor_7', 'sensor_8']
                pressure_data['time_stamp'] = pressure_data['time_stamp'].map(lambda x: x.lstrip('[').rstrip(''))
                pressure_data['time_stamp'] = pressure_data['time_stamp'].map(lambda x: x.lstrip(']').rstrip(''))
                pressure_data['time_stamp'].replace(regex=True,inplace=True,to_replace=r']',value=r'')
                pressure_data['time_stamp'] = pd.to_datetime(pressure_data['time_stamp'])
                pressure_data.index = pressure_data['time_stamp']
                del pressure_data['time_stamp']
                print(pressure_data.head())
                sensor_1 = pressure_data.sensor_1
                sensor_2 = pressure_data.sensor_2
                sensor_3 = pressure_data.sensor_3
                sensor_4 = pressure_data.sensor_4
                sensor_5 = pressure_data.sensor_5
                sensor_6 = pressure_data.sensor_6
                sensor_7 = pressure_data.sensor_7
                sensor_8 = pressure_data.sensor_8
                avgSensor1 = sensor_1.groupby(sensor_1.index.time).mean()
                avgSensor2 = sensor_2.groupby(sensor_2.index.time).mean()
                avgSensor3 = sensor_3.groupby(sensor_3.index.time).mean()
                avgSensor4 = sensor_4.groupby(sensor_4.index.time).mean()
                avgSensor5 = sensor_5.groupby(sensor_5.index.time).mean()
                avgSensor6 = sensor_6.groupby(sensor_6.index.time).mean()
                avgSensor7 = sensor_7.groupby(sensor_7.index.time).mean()
                avgSensor8 = sensor_8.groupby(sensor_8.index.time).mean()
                avgSensor1.plot(label="Sensor 1", legend=True)
                avgSensor2.plot(label="Sensor 2", legend=True)
                avgSensor3.plot(label="Sensor 3", legend=True)
                avgSensor4.plot(label="Sensor 4", legend=True)
                avgSensor5.plot(label="Sensor 5", legend=True)
                avgSensor6.plot(label="Sensor 6", legend=True)
                avgSensor7.plot(label="Sensor 7", legend=True)
                avgSensor8.plot(label="Sensor 8", legend=True)
                fig = plt.gcf()
                #plt.yscale('symlog')
                fig.set_size_inches(20,5)
                plt.title(filename)
                plt.show()
        os.chdir('..')
```
